# 🎯 Ejercicios: Funciones con Parámetros Mixtos

Este notebook contiene **5 ejercicios prácticos** para practicar funciones que combinan parámetros normales con `*args` o `**kwargs`.

## 🎯 Objetivos de aprendizaje:
- Combinar parámetros normales con `*args` en funciones
- Combinar parámetros normales with `**kwargs` en funciones
- Aplicar estos conceptos en situaciones prácticas

## 💡 Recordatorio:
- **Parámetro normal + *args**: `funcion(param_normal, *args)`
- **Parámetro normal + **kwargs**: `funcion(param_normal, **kwargs)`
- Los parámetros normales son obligatorios
- `*args` permite argumentos posicionales adicionales
- `**kwargs` permite argumentos con nombre adicionales
